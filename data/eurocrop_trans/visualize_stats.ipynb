Visualize Translation Stats

```python
from pathlib import Path
import pandas as pd

files = Path("/Users/ayshahchan/Desktop/PhD/eurocrop_trans").glob("*trans_stats.csv")

dfs = []

for f in files:
    df = pd.read_csv(f)
    
    # # Get first 2 characters of filename
    # prefix = f.name[:2]
    
    # # Add as new column
    # df["id"] = prefix
    
    dfs.append(df)


```

```python
dfs[0]
```

```python
for i in range(len(dfs)):
    col = dfs[i]["translation_flag"]

    country_code = dfs[i]["nuts"].iloc[0]
    # Drop missing values (NaN / None)
    valid = col.dropna()

    # Percentage of 1s
    percentage_right = (valid == 1).mean() * 100

    print(f"{country_code.upper()} Correctly translated: {percentage_right:.2f}%")

    col = dfs[i]["type_error"]

    # Drop missing values (NaN / None)
    valid = col.dropna()

    # Percentage of 1s
    percentage_translator_error = (valid == 1).mean() * 100


    # Drop missing values (NaN / None)
    valid = col.dropna()

    # Percentage of 2s
    percentage_weird_translation = (valid == 2 ).mean() * 100


    # Drop missing values (NaN / None)
    valid = col.dropna()

    # Percentage of 3s
    percentage_weird_and_no_agreement = (valid == 3 ).mean() * 100

    percentage_weird = percentage_weird_translation + percentage_weird_and_no_agreement

    print(f"{country_code.upper()} Machine translator error (one MT gives wrong translation): {percentage_translator_error:.2f}%")

    col = dfs[i]['Niche/uncommon name/spelling error/abbreviation']
    # Drop missing values (NaN / None)
    valid = col.dropna()

    # Percentage of 1s
    percentage_niche = (valid == 1 ).mean() * percentage_weird


    # Drop missing values (NaN / None)
    valid = col.dropna()

    # Percentage of 2s
    percentage_uncommon_name = (valid == 2 ).mean()  * percentage_weird



    # Percentage of 3s
    percentage_spelling_error = (valid == 3 ).mean() * percentage_weird


    # Drop missing values (NaN / None)
    valid = col.dropna()

    # Percentage of 4s
    percentage_abbreviation = (valid == 4 ).mean() * percentage_weird


    # Drop missing values (NaN / None)
    valid = col.dropna()

    # Percentage of 3s
    percentage_separable_prefix = (valid == 5 ).mean()  * percentage_weird


    print(f"{country_code.upper()} Niche crop name or Polysemy: {percentage_niche:.2f}%")
    print(f"{country_code.upper()} Colloquial name/Regional name: {percentage_uncommon_name:.2f}%")
    print(f"{country_code.upper()} Spelling error: {percentage_spelling_error:.2f}%")
    print(f"{country_code.upper()} Abbreviation/truncation in descriptions: {percentage_abbreviation:.2f}%")
    print(f"{country_code.upper()} Separable prefix: {percentage_separable_prefix:.2f}%")
```
